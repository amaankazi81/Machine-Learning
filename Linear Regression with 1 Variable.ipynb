```python
import pandas as pd                                      #importing liabraries
import seaborn as sns                                    #for visualizing data
from sklearn import linear_model                         #to load linear regression model
```

```python
df = pd.read_csv('agprm.csv')                             #by using pandas loading dataset
```

```python
df
```

```python
sns.lmplot(x='age', y="premium", data=df)                #visualizing dataset
```

```python
lr = linear_model.LinearRegression()                   #loading linear regression model
```

```python
lr.fit(df[['age']],df['premium'])                       #fitting with variables x and y where x is age and y is premium
```

```python
lr.predict([[21]])                               #Prediction for 21 years age premium
```

```python
lr.predict([[50]])                               #Prediction for 50 years age premium
```

```python
lr.coef_                                        #coefficient c for y=mx+c 
```

```python
lr.intercept_                                  #slope m for y=mx+c
```
